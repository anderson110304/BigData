```python
from sklearn.datasets import load_wine
```

# Изучение данных

```python
wine_dataset = load_wine()
```

```python
print("wine_dataset's keys: \n{}.\n".format(wine_dataset.keys()))
```

```python
print(wine_dataset['DESCR'] + '\n...')
```

```python
print("Сорта вина: {}".format(wine_dataset['target_names']))
```

```python
print("Признаки сортов вина: {}".format(wine_dataset['feature_names']))
```

```python
print("Тип массива data: {}".format(type(wine_dataset['data'])))
print("Форма массива data: {}".format(wine_dataset['data'].shape))
print("Первые 7 строк массива data:\n{}".format(wine_dataset['data'][:7]))
```

```python
print("Тип массива target: {}".format(type(wine_dataset['target'])))
print("Форма массива target: {}".format(wine_dataset['target'].shape))
print("Ответы:\n{}".format(wine_dataset['target'][:150])) 
print("\n0 - class_0\n1 - class_1\n2 - class_2")
```

# Разделение данных на обучающий и тестовый наборы

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(wine_dataset['data'], wine_dataset['target'], random_state=0)
```

```python
print("Форма массива x_train: {}".format(x_train.shape))
print("Форма массива y_train: {}".format(y_train.shape))
print("Форма массива x_text: {}".format(x_test.shape))
print("Форма массива y_text: {}".format(y_test.shape))
```

# Исследование данных с помощью визуализации

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
wine_dataframe = pd.DataFrame(x_train, columns=wine_dataset.feature_names)
plt.figure(figsize=(10, 8))
sns.heatmap(wine_dataframe.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

# Построение модели по методу k-ближайших средних
# Тестирование на произвольном наборе данных
# Получите прогноз модели
# Оцените качество модели

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train, y_train)
```

```python
x_new = np.array([[1.316e+01, 2.360e+00, 2.670e+00, 1.860e+01, 1.010e+02, 2.800e+00, 3.240e+00, 3.000e-01, 2.810e+00, 5.680e+00, 1.030e+00, 3.170e+00, 1.185e+03]])
print("Форма массива x_new: {}".format(x_new.shape))
```

```python
prediction = knn.predict(x_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированная метка: {}".format(wine_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(x_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred))
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(x_test, y_test)))
```
