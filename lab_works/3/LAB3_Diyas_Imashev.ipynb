# Визуализация данных с помощью matplotlib. Создание графиков с использованием «научной графики»

## 2. Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
import matplotlib.pyplot as plt
import numpy as np

a = 2
b = -3
c = 1

x_quad_func = np.arange(-5, 5, 0.1)
y_quad_func = a * (x_quad_func ** 2) + b * x_quad_func + c

plt.figure(figsize=(8, 6))
plt.plot(x_quad_func, y_quad_func, label=f'$f(x) = {a}x^2 + {b}x + {c}$')
plt.title('График квадратичной функции')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.grid(True)
plt.legend()

plt.show()
```

## 3. Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
import matplotlib.pyplot as plt
import numpy as np

A = 1.0        # Амплитуда
B = 0.5        # Частота
C = np.pi/2    # Фаза
D = 0.0        # Смещение вверх

x_harm_func = np.arange(0, 4 * np.pi, 0.1)
y_harm_func = A * np.cos(B * x_harm_func + C) + D

plt.figure(figsize=(8, 6))
plt.plot(x_harm_func, y_harm_func, label=f'$f(x) = {A} \cos({B}x + {C}) + {D}$')
plt.title('График гармонической функции (косинусная)')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.grid(True)
plt.legend()

plt.show()
```

## 4. Вывести графики п.2 и 3 в разные поля при помощи функции subplot()


```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_quad_func, y_quad_func)
plt.title('График квадратичной функции')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x_harm_func, y_harm_func)
plt.title('График гармонической функции (косинусная)')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.grid(True)

plt.tight_layout()

plt.show()
```

## 5. Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))

plt.plot(x_quad_func, y_quad_func, label='Квадратичная функция: $f(x) = ax^2 + c$')
plt.plot(x_harm_func, y_harm_func, label='Гармоническая функция: $f(x) = A \cos(Bx + C) + D$')

plt.title('Графики квадратичной и гармонической функций')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.grid(True)
plt.legend()

plt.show()
```

## 6. Построить график функции, заданной параметрически x(t)=cos⁡〖(3t),y=sin⁡〖(2t),t∈[0,4π]〗 〗

```python
def parametric_equations(t):
    x = np.cos(3 * t)
    y = np.sin(2 * t)
    return x, y

t_values = np.arange(0, 4 * np.pi, 0.01)
x_values, y_values = parametric_equations(t_values)

plt.plot(x_values, y_values, label='Параметрическая функция: $x(t) = \cos(3t), y(t) = \sin(2t)$')

plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.title('График параметрической функции')
plt.grid(True)
plt.legend()

plt.show()
```

## 7. Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
min_expenses = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]


# Строим столбиковую диаграмму
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(years, min_expenses, color='skyblue')
ax.set_ylim(bottom=0, top=max(min_expenses) + 1000)
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум (сом)')
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')

#текстовые аннотации
for i, value in enumerate(min_expenses):
    ax.text(years[i], value + 500, str(value), ha='center', va='bottom')

plt.show()
```

## 8. Построить круговой график по данным п. 7

```python
plt.pie(min_expenses, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Прожиточный минимум в КР (2012-2022)')

plt.show()
```

## 9. Построить график произвольной каркасной поверхности.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

a = 6378.137  
b = 6356.752  

phi = np.linspace(0, 2 * np.pi, 100)
theta = np.linspace(0, np.pi, 50)
phi, theta = np.meshgrid(phi, theta)

x = a * np.sin(theta) * np.cos(phi)
y = a * np.sin(theta) * np.sin(phi)
z = b * np.cos(theta)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(x, y, z, color='b', linewidth=0.5)

ax.set_xlabel('X (км)')
ax.set_ylabel('Y (км)')
ax.set_zlabel('Z (км)')
ax.set_title('Каркасная поверхность земного эллипсоида')

plt.show()
```
